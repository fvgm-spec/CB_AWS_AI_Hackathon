### Scraping postal codes of Germany

This code scraps the postal codes of Germany from [the Wikipedia page](https://simple.wikipedia.org/wiki/Postal_codes_in_Germany), then parse it as a json file and last store it in a csv file.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
```

```python
url = 'https://simple.wikipedia.org/wiki/Postal_codes_in_Germany'
```

```python
# make HTTP GET request to base URL
r = requests.get(url)
html = r.content
soup = BeautifulSoup(html, "html.parser")
```

```python
# parse response
content = BeautifulSoup(r.text, 'lxml')
```

```python
postcodes = []

for ul in content.find_all('ul')[100:]:
    print(ul.text)
```

```python
# Extracting state names
states = []
h2_element = soup.find_all('h2')
for h2 in h2_element:
    #Appending each h2 element to states list
    states.append(h2.text)
print(f"There are {len(states)} states in the web page")
states
```

```python
# Extracting each of the region names and code ranges
region_codes = []
h3_element = soup.find_all('h3')
for h3 in h3_element:
    
    region_codes.append(h3.text)

region_codes
```
